Pandas is a Python library used for data analysis and data manipulation.
It provides two main data structures:

Series → one-dimensional data (like one column)

DataFrame → two-dimensional table (like Excel spreadsheet)



```python
import pandas as pd
```

```python
#structure of series
series = pd.Series([1,2,3.0,4,5]) #its datatype preferece is object and float and int
#As per preference it will typecast
print(series)
```

```python
se = pd.Series(["a","c","srinu","vasu"])
print(se)
print(type(se))
```

```python
#Creating a series from numpy ndarray
#Only applicable for 1D array
import numpy as np
arr = np.array([12,23,65,32])
se = pd.Series(arr)
print(se)
```

```python
#Data acessing using index
se = pd.Series([1,2,3,4,5,6])
print(se[2],"\n")
print(se[1:],"\n")
print(se[[1,4]])#Acessing first and fourth index
```

```python
#Adding index manually
s = pd.Series([1,2,3,4,5,6],index=['a','b','c','d','e','f'])
print(s)
print(s['a'])
print(s['b':])
```

```python
print(s[['a','c','f']])
```

PANDAS DATAFRAME

```python
#Creating a dataframe b using dictionary
data = {"name ": ['srinu','vasu','charan'],
        "roll": [1,2,3],
        "marks": [69,78,96]}
df = pd.DataFrame(data)
df.index = range(1, len(df) + 1)
df
```

```python
data = {"name ": ['srinu','vasu','charan'],
        "roll": [1,2,3],
        "marks": [69,np.nan,96]}#null   NaN=not an number
df = pd.DataFrame(data,index=[1,2,3])
df
```

```python
#creating dataframe with tuple
data = [('1/1/2025',14,5,'Rain'),
        ('2/1/2025',13,6,'sun'),
        ('3/1/2025',14,3,'sun'),
        ('4/1/2025',12,8,'cloudy')]
df = pd.DataFrame(data,columns=['data','temperature','windspeed','climate condition'])
df
```
